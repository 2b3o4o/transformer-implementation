```python
import random

def generate_addition_problems(num_problems):
    problems = []
    for _ in range(num_problems):
        a = random.randint(0, 9)
        b = random.randint(0, 9)
        sum_ab = a + b
        problems.append(f"{a} + {b} = {sum_ab}")
    return problems

def write_to_file(problems, file_name):
    with open(file_name, 'w') as f:
        for problem in problems:
            f.write(f"{problem}\n")

num_problems = 1000  # Number of addition problems to generate
file_name = "../addition_problems-test.txt"  # File to save the problems
problems = generate_addition_problems(num_problems)
write_to_file(problems, file_name)
print(f"{num_problems} addition problems have been written to {file_name}")
```
