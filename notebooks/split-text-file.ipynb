```python
def split_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    max_length = 800000  # Max length of each split file
    start = 0  # Starting index of the substring
    part_num = 1  # Counter for file parts

    while start < len(content):
        # Find the last whitespace character within the limit
        end = min(start + max_length, len(content) - 1)
        
        # Adjust end index to not split in the middle of a word
        while end > start and content[end] not in (' ', '\n', '\t'):
            end -= 1

        # Extract substring
        substring = content[start:end]

        # Create new file and write substring
        output_file_path = f"../data/{file_path.split('.')[0]}_part{part_num}.txt"
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(substring)

        print(f"Created file: {output_file_path}")

        # Update start index and part number
        start = end + 1  # Skip the whitespace
        part_num += 1
```

```python
file_path = "../data/shakespeare.txt"
split_file(file_path)
```
